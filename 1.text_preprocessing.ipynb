```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Initialize the NLTK Porter Stemmer and get a list of English stop words
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Tokenize the text into words
    words = word_tokenize(text)
    
    # Remove punctuation and convert words to lowercase
    filtered_words = []
    for word in words:
        if word.isalpha():
            filtered_words.append(word.lower())
    
    # Remove stop words and apply stemming
    processed_words = []
    for word in filtered_words:
        if word not in stop_words:
            processed_words.append(stemmer.stem(word))

    # Join the processed words back into a text
    processed_text = ' '.join(processed_words)

    return processed_text
```

```python
# Specify the path to the text file
file_path = "file.txt"

# Read text from the file
with open(file_path, 'r') as file:
    text = file.read()

# Perform text preprocessing
processed_text = preprocess_text(text)
print(processed_text)
```
